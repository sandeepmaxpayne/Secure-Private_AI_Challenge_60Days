```
BASE = 10
PRECESION = 4
Q = 23784577778754875757458723
```

```
def encode(x_dec):
  return int(x_dec * (BASE * PRECESION)) % Q
```

```
def decode(x_fp):
  return (x_fp if x_fp <= Q/2 else x_fp - Q) / BASE**PRECESION
```

```
encode(0.5)
```

```
decode(27454844644644245123)
```

```
decode(5000 * 5000)
```
