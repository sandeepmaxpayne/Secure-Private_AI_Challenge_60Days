```
import random
```

```
Q = 277874548487425454754
```

```
x = 5
```

```
def encrypt(x, n_shares=1):
  n_shares = 3
  shares = list()

  for i in range(n_shares-1):
    shares.append(random.randint(0,Q))

  final_shares = Q - (sum(shares)%Q) + x
  shares.append(final_shares)
  return tuple(shares)
  
```

```
encrypt(5, n_shares=10)
```

```
def decrypt(shares):
  return sum(shares) % Q
```

```
def add(a, b):
  
  c = list()
  
  assert(len(a) == len(b))
  
  for i in range(len(a)):
    c.append((a[i] + b[i]) % Q)
    
  return tuple(c)
```

```
decrypt(add(encrypt(5), encrypt(10)))
```
